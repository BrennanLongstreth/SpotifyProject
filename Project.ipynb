### Hypothesis Tests To Clarify the difference in listening numbers from earlier graphs

Null Hypothesis: The mean of dance pop song lengths and the mean of hip hop song lengths are the same.  
Research Hypothesis: The mean of dance pop song lengths is significantly different from the mean of dance song lengths.  
Level of significance: 0.05  

```python
dataDancePop = sampleDF.loc[sampleDF[('genre')] == "dance pop"]['msPlayed']
dataHipHop = sampleDF.loc[sampleDF[('genre')] == "hip hop"]['msPlayed']
t,p = stats.ttest_ind(dataDancePop,dataHipHop)
print('t =',t,'p =',p)
```

Conclusion: We reject the null hypothesis because p is much lower than our level of significance. We accept the opposite of our Research hypothesis because our t value is negative.
  
This means that our listening time graph (at least the first 2 bars) are even more significantly different because the average hiphop song length is longer than the average dance pop song length.

### Second hypothesis test
Null Hypothesis: The mean of hip hop song lengths and the mean of electro pop song lengths are the same.  
Research Hypothesis: The mean of hip hop song lengths is significantly different from the mean of electro pop song lengths.  
Level of significance: 0.05 

```python
dataHipHop = sampleDF.loc[sampleDF[('genre')] == "hip hop"]['msPlayed']
dataElectroPop = sampleDF.loc[sampleDF[('genre')] == "electropop"]['msPlayed']
t,p = stats.ttest_ind(dataHipHop,dataElectroPop)
print('t =',t,'p =',p)
```

Conclusion: We reject the null hypothesis because p is much lower than our level of significance. We accept the opposite of our Research hypothesis because our t value is negative.
  
The mean length of Hip Hop songs is greater than the mean length of electro pop songs.
